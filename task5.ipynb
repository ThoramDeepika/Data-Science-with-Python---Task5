```python
# 1. How many have heart disease and how many people doesn't have haert disesase?
data.target.value_counts()
```

```python
data.target.value_counts().plot(kind = "bar", color=["orchid","salmon"])
plt.title("Heart disease values")
plt.xlabel("1 = Heart disease, 0= No Heart disease")
plt.ylabel("Amount");
```

```python
#2. People of which sex has most heart disease?
pd.crosstab(data.target,data.sex)
```

```python
sns.countplot(x= "target", data=data, hue= "sex")
plt.title("Heart disease frequency for sex")
plt.xlabel("0 = No heart disease, 1 = Heart disease");
```

```python
#3. People of which sex has which type of chest pain most?
data.cp.value_counts()
```

```python
data.cp.value_counts().plot(kind = "bar",color = ["salmon", "lightskyblue", "springgreen","khaki"])
```

```python
pd.crosstab(data.sex,data.cp)
```

```python
pd.crosstab(data.sex,data.cp).plot(kind= "bar", color = ["coral","lightskyblue","plum","khaki"])
plt.title("Type of chest pain for sex")
plt.xlabel("0 = Female, 1 = Male");
```

```python
#4. People with chest pain are most pron to have heart disease?
pd.crosstab(data.cp,data.target)
```

```python
sns.countplot(x="cp", data = data, hue= "target");
```

```python
sns.displot(x="age", data = data, bins = 30, kde= True);
```

```python
sns.displot(x="thalach", data = data, bins = 30, kde = True, color = "chocolate");
```

```python
#5. What is the distribution of patients' ages in the dataset?
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=20, kde=True)
plt.title('Distribution of Patients\' Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#6. Plot the correlation between age and maximum heart rate achieved
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='thalach', data=data)
plt.title('Correlation Between Age and Maximum Heart Rate Achieved')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate Achieved (thalach)')
plt.grid(True)
plt.show()
```

```python
#7. Plot the distribution of chest pain types between patients with and without heart disease
plt.figure(figsize=(10, 6))
sns.countplot(x='cp', hue='target', data=data)
plt.title('Distribution of Chest Pain Types Between Patients With and Without Heart Disease')
plt.xlabel('Chest Pain Type (cp)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.grid(True)
plt.show()
```

```python
#8. Plot the relationship between serum cholesterol and the presence of heart disease
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='chol', data=data)
plt.title('Relationship Between Serum Cholesterol and Presence of Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Serum Cholesterol (chol)')
plt.xticks([0, 1], ['No', 'Yes'])
plt.grid(True)
plt.show()
```
